# Pre-sort images
This notebook pre-sorts input images to remove extreme values, before passing data into ML model.

## Compute average pixel value of different classes

# 1. Resize images

```python
# resize images to 64x64
from PIL import Image, ImageOps
import os
def resize_multiple_images(src_path, dst_path):
    # Here src_path is the location where images are saved.
    for filename in os.listdir(src_path):
        try:
            img=Image.open(src_path+filename)
            new_img = img.resize((80,80))
            if not os.path.exists(dst_path):
                os.makedirs(dst_path)
            new_img.save(dst_path+filename)
            # print('Resized, grayed, and saved {} successfully.'.format(filename))
        except:
            continue



src_path = "pre_sort_images/full_size/FMGP/"
dst_path = "pre_sort_images/resized/FMGP/"
resize_multiple_images(src_path, dst_path)

src_path = "pre_sort_images/full_size/buildings/"
dst_path = "pre_sort_images/resized/buildings/"
resize_multiple_images(src_path, dst_path)

src_path = "pre_sort_images/full_size/compasses/"
dst_path = "pre_sort_images/resized/compasses/"
resize_multiple_images(src_path, dst_path)

src_path = "pre_sort_images/full_size/other/"
dst_path = "pre_sort_images/resized/other/"
resize_multiple_images(src_path, dst_path)

src_path = "pre_sort_images/full_size/industrial/"
dst_path = "pre_sort_images/resized/industrial/"
resize_multiple_images(src_path, dst_path)

src_path = "pre_sort_images/full_size/turntables/"
dst_path = "pre_sort_images/resized/turntables/"
resize_multiple_images(src_path, dst_path)

src_path = "pre_sort_images/full_size/cisterns/"
dst_path = "pre_sort_images/resized/cisterns/"
resize_multiple_images(src_path, dst_path)
```

# Input images to array and find mean and std of pixel values

Using the red channel of the image, as it has the greatest variation

```python
import cv2
from PIL import Image
import os
import numpy as np
import re

def to_dataframe(mean,std):
    m = pd.DataFrame(mean)
    s = pd.DataFrame(std)
    df = pd.concat([m,s], axis=1)
    df.columns = ['mean', 'std']
    return df

def get_data(path):
    all_means=[]
    all_std=[]
    for filename in os.listdir(path):
        if "jpg" in filename:
            try:
                img=cv2.imread(path + filename)
                # img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
                img = img[:,:,2]
                mean = np.mean(img)
                std = np.std(img)
                all_means.append(mean)
                all_std.append(std)
            except:
                continue
    return to_dataframe(np.array(all_means), np.array(all_std))


FMGP = "pre_sort_images/resized/FMGP/"
compasses = "pre_sort_images/resized/compasses/"
buildings = "pre_sort_images/resized/buildings/"
other = "pre_sort_images/resized/other/"
industrial = "pre_sort_images/resized/industrial/"
turntables = "pre_sort_images/resized/turntables/"
cisterns = "pre_sort_images/resized/cisterns/"

F = get_data(FMGP)
C = get_data(compasses)
B = get_data(buildings)
O = get_data(other)
I = get_data(industrial)
T = get_data(turntables)
W = get_data(cisterns)

def summary(df):
    for column in df: 
        print(column)
        print("Mean: ", np.mean(df[column]))
        print("Min: ", np.min(df[column]))
        print("Max: ", np.max(df[column]))
        print('1%: ', np.percentile(df[column],1))
        print('99%:', np.percentile(df[column],99))
        print('\n')
    return

           
print('Summary statistics of pixel standard deviation: ')
print('\nFMGP')
summary(F)

print('\nCompasses')
summary(C)


print('\nBuildings')
summary(B)


print('\nWater')
summary(W)

```

```python
# now, let's apply a filter
# remove all items that are beyond the 1% and 99% of FMGP pixel mean and std dev

def cutoff(df):
    df['cutoff'] = 1
    df.loc[
        (df['mean'] < np.percentile(F['mean'],1)) |
        (df['mean'] > np.percentile(F['mean'],99)) |
        (df['std'] < np.percentile(F['std'],1)) |
        (df['std'] > np.percentile(F['std'],99)),
        'cutoff'] = 0
    return  

cutoff(C)
cutoff(B)
cutoff(W)

print('\nPercent compass circles remaining: ')
print(np.mean(C['cutoff']))

print('\nPercent building circles remaining: ')
print(np.mean(B['cutoff']))

print('\nPercent water circles remaining: ')
print(np.mean(W['cutoff']))
```
